```python
equiments = root.findall('Equipment')
```

```python
Equiments = []
for equiment in equiments:
    Equiment = {}
    #print(dir(equiment.keys()))
    if 'ID' in equiment.keys():
        index = equiment.keys().index('ID')
        Equiment['ID'] =equiment.values()[index]
    if 'TagName' in equiment.keys():
        index = equiment.keys().index('TagName')
        Equiment['TagName'] =equiment.values()[index] 
    
    if 'ComponentName' in equiment.keys():
        index = equiment.keys().index('ComponentName')
        Equiment['ComponentName'] =equiment.values()[index]
    
    if 'ComponentClass' in equiment.keys():
        index = equiment.keys().index('ComponentClass')
        Equiment['ComponentClass'] =equiment.values()[index]
    Nozzle = []
    for child in equiment.getchildren():
        #print(child.tag)
        if child.tag == 'Nozzle':
             if 'ID' in child.keys():
                index = child.keys().index('ID')
                Nozzle.append(child.values()[index])
    Equiment['Nozzles_ID'] = Nozzle
    Equiments.append(Equiment)

#print(Equiments)

df_equiment = pd.DataFrame(Equiments)
```

```python
pipline_netseg = root.findall('PipingNetworkSystem/PipingNetworkSegment')
```

```python
PipNets = []
for pip_net in pipline_netseg:
    PipNet = {}
    if 'ID' in pip_net.keys():
        index = pip_net.keys().index('ID')
        PipNet['ID'] =pip_net.values()[index]
    if 'TagName' in pip_net.keys():
        index = pip_net.keys().index('TagName')
        PipNet['TagName'] =pip_net.values()[index] 
    for i in pip_net:
        if i.tag == 'Connection':
            #print(i.keys())
            #print(i.values())
            if 'FromID' in i.keys():
                index = i.keys().index('FromID')
                PipNet['Connection_From_ID'] =i.values()[index]
                #print(i.values()[index])
                element  = root.findall('''.//*[@ID='{}']'''.format(PipNet['Connection_From_ID']))
                #print(len(element))
                if len(element) >= 1:
                    PipNet['Connection_From_Comp'] = element[0].tag
            if 'ToID' in i.keys():
                index = i.keys().index('ToID')
                PipNet['Connection_To_ID'] =i.values()[index]
                element  = root.findall('''.//*[@ID='{}']'''.format(PipNet['Connection_To_ID']))
                if len(element)>=1:
                    PipNet['Connection_To_Comp'] = element[0].tag
        if i.tag == 'GenericAttributes':
            for j in i:
                if j.values()[0] == 'ActualFrom':
                    PipNet['ActualFrom'] = j.values()[2]
                if j.values()[0]== 'ActualTo':
                    PipNet['ActualTo'] = j.values()[2]
                
                if j.values()[0]=='Area':
                    PipNet['Area'] = j.values()[2]
    PipNets.append(PipNet)
    
df_pip_seg = pd.DataFrame(PipNets)

        
```

```python
pf_comp = []
for i in root.findall('ProcessInstrumentationFunction'):
    for j in i :
        if j.tag == 'ActuatingFunction' or j.tag == 'InformationFlow' or j.tag =='ProcessSignalGeneratingFunction':
            Pf_comp = {}
            Pf_comp['Comp_type'] = j.tag
            if 'ID' in j.keys():
                index = j.keys().index('ID')
                Pf_comp['ID'] =j.values()[index]
            if 'TagName' in j.keys():
                index = j.keys().index('TagName')
                Pf_comp['TagName'] =j.values()[index] 
            for k in j:
                if k.tag == 'Association': 
                    Pf_comp = {}
                    Pf_comp['Comp_type'] = j.tag
                    if 'ID' in j.keys():
                        index = j.keys().index('ID')
                        Pf_comp['ID'] =j.values()[index]
                    if 'TagName' in j.keys():
                        index = j.keys().index('TagName')
                        Pf_comp['TagName'] =j.values()[index] 
                    if 'ItemID' in k.keys():
                        index = k.keys().index('ItemID')
                        Pf_comp['ChildID'] =k.values()[index]
                        element  = root.findall('''.//*[@ID='{}']'''.format(Pf_comp['ChildID']))
                        Pf_comp['Child_type'] = element[0].tag
                    if 'TagName' in k.keys():
                        index = k.keys().index('TagName')
                        Pf_comp['Child_TagName'] =k.values()[index]
                    if 'Type' in k.keys():
                        index = k.keys().index('Type')
                        Pf_comp['Child_Relation_Type'] =k.values()[index]
                    pf_comp.append(Pf_comp)

                if k.tag == 'GenericAttributes':
                    Pf_comp = {}
                    Pf_comp['Comp_type'] = j.tag
                    if 'ID' in j.keys():
                        index = j.keys().index('ID')
                        Pf_comp['ID'] =j.values()[index]
                    if 'TagName' in j.keys():
                        index = j.keys().index('TagName')
                        Pf_comp['TagName'] =j.values()[index] 
                    for child in k:
                        if child.values()[0] == 'SIGNALLINETYPE':
                            Pf_comp['SIGNALLINETYPE'] = child.values()[2]
                    pf_comp.append(Pf_comp)

                if k.tag == 'Connection':
                    Pf_comp = {}
                    Pf_comp['Comp_type'] = j.tag
                    if 'ID' in j.keys():
                        index = j.keys().index('ID')
                        Pf_comp['ID'] =j.values()[index]
                    if 'TagName' in j.keys():
                        index = j.keys().index('TagName')
                        Pf_comp['TagName'] =j.values()[index] 
                    if 'FromID' in k.keys():
                        index = k.keys().index('FromID')
                        Pf_comp['Connection_From_ID'] =k.values()[index]
                        #print(i.values()[index])
                        element  = root.findall('''.//*[@ID='{}']'''.format(Pf_comp['Connection_From_ID']))
                        #print(len(element))
                        if len(element) >= 1:
                            Pf_comp['Connection_From_Comp'] = element[0].tag
                    if 'ToID' in k.keys():
                        index = k.keys().index('ToID')
                        Pf_comp['Connection_To_ID'] =k.values()[index]
                        element  = root.findall('''.//*[@ID='{}']'''.format(Pf_comp['Connection_To_ID']))
                        if len(element)>=1:
                            Pf_comp['Connection_To_Comp'] = element[0].tag
                    pf_comp.append(Pf_comp)
    
```
